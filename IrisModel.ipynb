# Import

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.tree import plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

# Loading the Dataset

```python
iris=sns.load_dataset('iris')
iris
```

## Head of Dataset

```python
iris.head(10)
```

## Visualization

```python
sns.set_style('darkgrid')
sns.FacetGrid(iris,hue="species").map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

## Separate the dependent value

```python

Y=iris['species']
Y
```

## Correlation between all of them

```python
iris.corr()
```

## Separate the independent value

```python

X=iris.drop(['species'],axis=1)
X
```

## Encodes the dependent variable

```python
#creating dependent data 
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

## Distribute the training and testing data

```python
#creating training and testing data
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
```

# Logistic Regression

```python
lr=LogisticRegression(solver='newton-cg')
```

```python
#Creating Model ofLogistic Regression
lr.fit(xtrain,ytrain)
```

```python
#Predicting Model ofLogistic Regression
ped1=lr.predict(xtest)
ped1
```

### Confusion matrix for Logistic Regression

```python
#Confusion Matrix of Logistic Regression
confusion_matrix(ytest,ped1)
```

### Accuracy Score for Logistic Regression

```python
#Accuracy Score of Logistic Regression
z=accuracy_score(ytest,ped1)
z
```

# Decision Tree

```python
#Creating Model of  Decision Tree
dtree=DecisionTreeClassifier()
dtree.fit(xtrain,ytrain)
```

```python
plt.figure(figsize=(10,20))
dec_tree = plot_tree(decision_tree=dtree, feature_names = iris.columns[:-1], 
                     class_names =["setosa", "vercicolor", "verginica"], filled = True )
```

```python
#Predicting Model of Decision Tree
ped2=dtree.predict(xtest)
ped2
```

### Confusion Matrix for Decision Tree

```python
#Confusion Matrix of Decision Tree 
confusion_matrix(ytest,ped2)
```

### Accuracy Score for Decision Tree

```python
#Accuracy Score of Decision Tree
z=accuracy_score(ytest,ped2)
z
```

# Random Forest

```python
clf=RandomForestClassifier(n_estimators=100)
```

```python
#Creating Model of Random Forest
clf.fit(xtrain,ytrain)
```

```python
#Predicting Model of Random Forest
ped3=clf.predict(xtest)
ped3
```

### Confusion Matrix for Random Forest

```python
#Confusion Matrix of Random Forest
confusion_matrix(ytest,ped3)
```

### Accuracy Score for Random Forest

```python
#Accuracy Score of Random Forest
z=accuracy_score(ytest,ped3)
z
```

# Result 

### In this model we saw that the accuracy of Decision tree is 2% more than Logistic Regression and Random Forest
